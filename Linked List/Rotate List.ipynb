# Rotate List
$\quad$ Given the `head` of a linked list, rotate the list to the right by `k` places.

# Examples
**Example 1:**<br>
Input: head = [1,2,3,4,5], k = 2<br>
Output: [4,5,1,2,3]



**Example 2:**<br>
Input: head = [0,1,2], k = 4<br>
Output: [2,0,1]



```python
from typing import Optional

# Definition for singly-linked list.


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def rotateRight(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        if head == None:
            return None

        result = head
        total = -1
        for i in range(k):
            head = head.next
            if head == None:
                total = i + 1
                break

        if total != -1:
            k = k % total

        head = result
        for i in range(k):
            head = head.next

        second = result
        while head.next != None:
            head = head.next
            second = second.next

        end = head
        end.next = result
        result = second.next
        second.next = None
        
        return result
```
