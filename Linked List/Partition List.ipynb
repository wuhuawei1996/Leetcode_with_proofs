# Partition List
$\quad$ Given the `head` of a linked list and a value `x`, partition it such that all nodes less than `x` come before nodes greater than or equal to `x`.

$\quad$ You should preserve the original relative order of the nodes in each of the two partitions.

# Examples
**Example 1:**<br>
Input: head = [1,4,3,2,5,2], x = 3<br>
Output: [1,2,2,4,3,5]



**Example 2:**<br>
Input: head = [2,1], x = 2<br>
Output: [1,2]

```python
from typing import Optional

# Definition for singly-linked list.


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def partition(self, head: Optional[ListNode], x: int) -> Optional[ListNode]:
        less = ListNode(0)
        more = ListNode(0)
        head_less = less
        head_more = more
        while head != None:
            if head.val < x:
                head_less.next = head
                head_less = head_less.next
            else:
                head_more.next = head
                head_more = head_more.next
            head = head.next

        head_less.next = more.next
        head_more.next = None
        
        return less.next
```
