# Remove Duplicates from Sorted List II
$\quad$ Given the `head` of a sorted linked list, delete all nodes that have duplicate numbers, leaving only distinct numbers from the original list. Return the linked list sorted as well.

# Examples
**Example 1:**<br>
Input: head = [1,2,3,3,4,4,5]<br>
Output: [1,2,5]



**Example 2:**<br>
Input: head = [1,1,1,2,3]<br>
Output: [2,3]



```python
from typing import Optional

# Definition for singly-linked list.


class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        result = ListNode(0.1, next=head)
        head = result
        while head != None and head.next != None:
            last = head
            head = head.next
            val = head.val
            count = 0
            while head != None and head.val == val:
                count += 1
                head = head.next
            if count > 1:
                last.next = head
                head = last
            else:
                head = last.next

        return result.next
```
