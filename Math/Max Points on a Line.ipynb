# Max Points on a Line
Given an array of `points` where <code>points[i] = [x<sub>i</sub>, y<sub>i</sub>]</code> represents a point on the X-Y plane, return the maximum number of points that lie on the same straight line.

```python
from math import gcd


class Solution:
    def get_line(self, p_1, p_2):
        x_1, y_1 = p_1
        x_2, y_2 = p_2
        A = y_1 - y_2
        B = x_2 - x_1
        C = y_2 * x_1 - x_2 * y_1
        d = gcd(A, B, C)
        A = A // d
        B = B // d
        C = C // d
        if A < 0 or (A == 0 and B < 0):
            A = -A
            B = -B
            C = -C
        return (A, B, C)

    def maxPoints(self, points: list[list[int]]) -> int:
        if len(points) <= 2:
            return len(points)
        lines = {}
        for i in range(len(points)):
            for j in range(i + 1, len(points)):
                line = self.get_line(points[i], points[j])
                if line not in lines:
                    lines[line] = set()
                lines[line].add(i)
                lines[line].add(j)

        return max([len(line) for line in lines.values()])
```

```python
from collections import defaultdict

class Solution:
    def maxPoints(self, points: list[list[int]]) -> int:
        if len(points) <= 2:
            return len(points)
        
        def find_slope(p1, p2):
            x1, y1 = p1
            x2, y2 = p2
            if x1-x2 == 0:
                return float('inf')
            return (y1-y2)/(x1-x2)
        
        ans = 1
        for i, p1 in enumerate(points):
            slopes = defaultdict(int)
            for j, p2 in enumerate(points[i+1:]):
                slope = find_slope(p1, p2)
                slopes[slope] += 1
            if len(slopes) > 0:
                ans = max(max(slopes.values()), ans)
                
        return ans+1
```
