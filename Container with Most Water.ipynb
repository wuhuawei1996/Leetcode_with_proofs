# Container with Most Water
$\quad$ You are given an integer array height of length n. There are n vertical lines drawn such that the two endpoints of the ith line are $(i, 0)$ and $(i, height[i])$.

$\quad$ Find two lines that together with the x-axis form a container, such that the container contains the most water.

$\quad$ Return the maximum amount of water a container can store.

$\quad$ Notice that you may not slant the container.

# Examples
**Example 1:**<br>
Input: $height = [1,8,6,2,5,4,8,3,7]$<br>
Output: $49$<br>
Explanation: The above vertical lines are represented by array $[1,8,6,2,5,4,8,3,7]$. In this case, the max area of water (blue section) the container can contain is $49$.


**Example 2:**<br>
Input: $height = [1,1]$<br>
Output: $1$
 

# Analysis

```python
class Solution:
    def maxArea(self, height: list[int]) -> int:
        max_area = 0
        left = 0
        right = len(height) - 1

        while left < right:
            max_area = max(max_area, (right - left) * min(height[left], height[right]))

            if height[left] < height[right]:
                left += 1
            else:
                right -= 1
        
        return max_area
```
