# N-Queens II

The n-queens puzzle is the problem of placing `n` queens on an `n x n` chessboard such that no two queens attack each other.

Given an integer `n`, return the number of distinct solutions to the n-queens puzzle.

## Examples

**Example 1:**


```
Input: n = 4
Output: 2
```
Explanation: There are two distinct solutions to the 4-queens puzzle as shown.

**Example 2:**
```
Input: n = 1
Output: 1
```

````python
class Solution:
    def totalNQueens(self, n: int) -> int:
        total = 0

        def backtrack(row, ban_cols, diag1, diag2):
            nonlocal total
            if row == n:
                total += 1
                return

            for col in range(n):
                if col in ban_cols or row + col in diag1 or row - col in diag2:
                    continue
                ban_cols.add(col)
                diag1.add(row + col)
                diag2.add(row - col)
                backtrack(row + 1, ban_cols, diag1, diag2)
                ban_cols.remove(col)
                diag1.remove(row + col)
                diag2.remove(row - col)

        backtrack(0, set(), set(), set())

        return total
````
