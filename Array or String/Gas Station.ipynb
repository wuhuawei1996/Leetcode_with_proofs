# Gas Station
$\quad$ There are $n$ gas stations along a circular route, where the amount of gas at the ith station is $gas[i]$.

$\quad$ You have a car with an unlimited gas tank and it costs $cost[i]$ of gas to travel from the ith station to its next $(i + 1)$-th station. You begin the journey with an empty tank at one of the gas stations.

$\quad$ Given two integer arrays gas and cost, return the starting gas station's index if you can travel around the circuit once in the clockwise direction, otherwise return $-1$. If there exists a solution, it is guaranteed to be unique.

# Examples
**Example 1:**<br>
Input: $gas = [1,2,3,4,5]$, $cost = [3,4,5,1,2]$<br>
Output: $3$<br>
Explanation:<br>
Start at station $3$ (index $3$) and fill up with $4$ unit of gas. Your tank = $0 + 4 = 4$.<br>
Travel to station $4$. Your tank = $4 - 1 + 5 = 8$.<br>
Travel to station $0$. Your tank = $8 - 2 + 1 = 7$.<br>
Travel to station $1$. Your tank = $7 - 3 + 2 = 6$.<br>
Travel to station $2$. Your tank = $6 - 4 + 3 = 5$.<br>
Travel to station $3$. The cost is $5$. Your gas is just enough to travel back to station $3$.<br>
Therefore, return $3$ as the starting index.

**Example 2:**<br>
Input: $gas = [2,3,4]$, $cost = [3,4,3]$<br>
Output: $-1$<br>
Explanation:
You can't start at station $0$ or $1$, as there is not enough gas to travel to the next station.<br>
Let's start at station $2$ and fill up with $4$ unit of gas. Your tank = $0 + 4 = 4$.<br>
Travel to station $0$. Your tank = $4 - 3 + 2 = 3$.<br>
Travel to station $1$. Your tank = $3 - 3 + 3 = 3$.<br>
You cannot travel back to station $2$, as it requires $4$ unit of gas but you only have $3$.<br>
Therefore, you can't travel around the circuit once no matter where you start.

# Analysis
$\quad$ Assume that $gas=[g_0,\cdots,g_{n-1}]$ and $cost=[c_0,\cdots,c_{n-1}]$. We prove that a solution exists if and only if $g_0+\cdots+g_{n-1}\ge c_0+\cdots+c_{n-1}$. It is clear that only the sufficiency part needs to be proved. We claim that, in this case, there exists $i\in[0,n-1]$ such that 
$$\begin{align}
&g_i\ge c_i,\notag\\
&g_i+g_{i+1}\ge c_i+c_{i+1},\\
&\cdots,\notag\\
&g_i+g_{i+1}+\cdots+g_{n-1}\ge c_i+c_{i+1}+\cdots+c_{n-1}.\notag
\end{align}$$
Otherwise, there exists a sequence $0\le i_1<i_2<\cdots<i_k<n-1$ such that 
$$\begin{align*}
&g_0+\cdots+g_{i_1}<c_0+\cdots+c_{i_1},\\
&g_{i_1+1}+\cdots+g_{i_2}<c_{i_1+1}+\cdots+c_{i_2},\\
&\cdots,\\
&g_{i_k+1}+\cdots+g_{n-1}<c_{i_k+1}+\cdots+c_{n-1}.
\end{align*}$$
Adding all the equations together, we get
$$g_0+\cdots+g_{n-1}<c_0+\cdots+c_{n-1},$$
which contradicts the assumption. Now assume that $i\in[0,n-1]$ is the smallest index such that the inequalities in (1) hold. If there exists $i'\in [0,i-1]$ such that 
$$\begin{align*}
&g_{i'}\ge c_{i'},\notag\\
&g_{i'}+g_{i'+1}\ge c_{i'}+c_{i'+1},\\
&\cdots,\notag\\
&g_{i'}+g_{i'+1}+\cdots+g_{i''}\ge c_{i'}+c_{i'+1}+\cdots+c_{i''},\notag
\end{align*}$$
for some $i''\ge\max(i',i-1)$, then
$$\begin{align*}
&g_{i'}\ge c_{i'},\notag\\
&g_{i'}+g_{i'+1}\ge c_{i'}+c_{i'+1},\\
&\cdots,\notag\\
&g_{i'}+g_{i'+1}+\cdots+g_{i-1}\ge c_{i'}+c_{i'+1}+\cdots+c_{i-1},\notag\\
&g_{i'}+g_{i'+1}+\cdots+g_{i}\ge c_{i'}+c_{i'+1}+\cdots+c_{i},\notag\\
&\cdots,\\
&g_{i'}+g_{i'+1}+\cdots+g_{n-1}\ge c_{i'}+c_{i'+1}+\cdots+c_{n-1},\notag\\
\end{align*}$$
which contradicts the minimality of $i$. Hence there exists $0\le i_1<i_2<\cdots<i_j\le i-1$ such that
$$\begin{align*}
&g_0\ge c_0,\\
&\cdots,\\
&g_{0}+\cdots+g_{i_1-1}\ge c_0+\cdots+c_{i_1-1},\\
&g_{0}+\cdots+g_{i_1}<c_0+\cdots+c_{i_1},\\
&g_{i_1+1}\ge c_{i_1+1},\\
&\cdots,\\
&g_{i_1+1}+\cdots+g_{i_2-1}\ge c_{i_1+1}+\cdots+c_{i_2-1},\\
&g_{i_1+1}+\cdots+g_{i_2}<c_{i_1+1}+\cdots+c_{i_2},\\
&\cdots,\\
&g_{i_j+1}\ge c_{i_j+1},\\
&\cdots,\\
&g_{i_j+1}+\cdots+g_{i-2}\ge c_{i_j+1}+\cdots+c_{i-2},\\
&g_{i_j+1}+\cdots+g_{i-1}<c_{i_j+1}+\cdots+c_{i-1},
\end{align*}$$
For any $s\in[0,i_1-1]$, since 
$$g_0+\cdots+g_{s}\ge c_0+\cdots+c_{s},$$
we have 
$$g_i+\cdots+g_{n-1}+g_0+\cdots+g_{s}\ge c_i+\cdots+c_{n-1}+c_0+\cdots+c_{s}.$$
Moreover, since 
$$\begin{align*}
&g_{i_1+1}+\cdots+g_{i_2}<c_{i_1+1}+\cdots+c_{i_2},\\
&g_{i_2+1}+\cdots+g_{i_3}<c_{i_2+1}+\cdots+c_{i_3},\\
&\cdots,\\
&g_{i_j+1}+\cdots+g_{i-1}<c_{i_j+1}+\cdots+c_{i-1},
\end{align*}$$
we have 
$$g_{i_1+1}+\cdots+g_{i-1}<c_{i_1+1}+\cdots+c_{i-1}.$$
Since 
$$g_0+\cdots+g_{n-1}\ge c_0+\cdots+c_{n-1},$$
we have 
$$g_i+\cdots+g_{n-1}+g_0+\cdots+g_{i_1}\ge c_i+\cdots+c_{n-1}+c_0+\cdots+c_{i_1}.$$
By mathematical induction, we can prove that for any $s\in[0,i-1]$, we have 
$$g_i+\cdots+g_{n-1}+g_0+\cdots+g_{s}\ge c_i+\cdots+c_{n-1}+c_0+\cdots+c_{s},$$
which means that we can start from station $i$.

$\quad$ Thus, the sufficiency is proven. Moreover, our proof actually provides a constructive method for obtaining the solution. Since the problem statement assumes the solution is unique, the solution constructed in this way is exactly the desired one.

```python
class Solution:
    def canCompleteCircuit(self, gas: list[int], cost: list[int]) -> int:
        if sum(gas) < sum(cost):
            return -1
                
        current_gas = 0
        start = 0
        for i in range(len(gas)):
            current_gas += gas[i] - cost[i]
            if current_gas < 0:
                current_gas = 0
                start = i + 1

        return start
```
